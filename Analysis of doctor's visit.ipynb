# Heatmap of Age and Income

```python
import numpy as np

# Bin age and income data
age_bins = np.linspace(df['age'].min(), df['age'].max(), 6)
income_bins = np.linspace(df['income'].min(), df['income'].max(), 6)

# Create age and income categories
df['age_category'] = pd.cut(df['age'], bins=age_bins, labels=['<20', '20-30', '30-40', '40-50', '50+'])
df['income_category'] = pd.cut(df['income'], bins=income_bins, labels=['Low', 'Low-Medium', 'Medium', 'Medium-High', 'High'])

# Create a pivot table for heatmap
heatmap_data = df.pivot_table(index='age_category', columns='income_category', values='reduced', aggfunc=np.mean)

# Plot the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='.2f', linewidths=0.5, annot_kws={"size": 10})
plt.xlabel('Income Category')
plt.ylabel('Age Category')
plt.title('Average Reduced Activity by Age and Income')
plt.show()
```

# Heatmap of Average Reduced Activity by Illness Type and Chronic Condition

```python
# Create a pivot table for heatmap
heatmap_data_reduced = df.pivot_table(index='illness', columns='lchronic', values='reduced', aggfunc=np.mean)

# Plot the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data_reduced, cmap='coolwarm', annot=True, fmt='.2f', linewidths=0.5, annot_kws={"size": 10})
plt.xlabel('Chronic Condition')
plt.ylabel('Illness Type')
plt.title('Average Reduced Activity by Illness Type and Chronic Condition')
plt.show()

```
